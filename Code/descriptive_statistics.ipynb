```python
import numpy as np
import pandas as pd

from scipy.stats import zscore
```

```python
original_data = pd.read_csv('Genshin_impact.csv',index_col=0,parse_dates=True)
original_data
```

```python
print(round(zscore(original_data).describe().T,2).to_latex())
```

```python
print(round(zscore(original_data).skew(),2).to_latex())
```

```python
print(round(zscore(original_data).kurt(),2).to_latex())
```

```python
original_data.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.heatmap(original_data.corr(),annot=True,vmax=True,square=True,cmap='Blues')
plt.show()
```
